```python
filename = "banquet_input_9.txt"

with open(filename, encoding="utf-8") as src:
    n, m = map(int, src.readline().split())
    ovp_pairs = [list(map(int, line.split())) for line in src.readlines()]

table_1 = {'persons': set(), 'black_list': set()}
table_2 = {'persons': set(), 'black_list': set()}

answer = "YES"
for pair in ovp_pairs:
    person_1, person_2 = pair

    lst = [28, 71]
    if person_1 in lst or person_2 in lst:
        print(person_1, person_2)
         
        if person_1 not in table_1['black_list'] and \
        person_2 not in table_2['black_list']:
            table_1['persons'].add(person_1)
            table_2['persons'].add(person_2)
            table_1['black_list'].add(person_2)
            table_2['black_list'].add(person_1)

        elif person_1 not in table_2['black_list'] and \
            person_2 not in table_1['black_list']:
                table_1['persons'].add(person_2)
                table_2['persons'].add(person_1)
                table_1['black_list'].add(person_1)
                table_2['black_list'].add(person_2)

        else:
            answer = "NO"

print(answer)
```
